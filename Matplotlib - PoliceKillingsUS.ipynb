**INSTRUCTION : Plot Most common 15 Name OR Surname of killed people** 

shape

```python
pairs = df['name'].str.split()
pairs
```

```python
pairs.shape
```

```python
pairs[0]
```

```python
names, surnames = zip(*pairs)
```

```python
names[0]
```

```python
name_surname= names + surnames
name_surname_counts = Counter(name_surname)
name_surname_counts
```

```python
name_surname_counts
```

```python
most_common_name_surname= name_surname_counts.most_common(15)
most_common_name_surname
```

```python
name_df=pd.DataFrame(most_common_name_surname, columns=['name_or_surname_most_frequency', 'frequency'] )
name_df
```

```python
name_df.plot(kind='bar', x= 'name_or_surname_most_frequency',y= 'frequency', figsize=(10,6))
plt.title('Most Common 15 Name or Surname of killed people')
plt.xlabel('Names or Surnames')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

for i, v in enumerate(name_df['frequency']):
    plt.text(i, v, str(v));
```

```python
# with pandas
```

```python
plt.figure(figsize=(8,6), dpi=100)
plt.bar(name_df['name_or_surname_most_frequency'],name_df['frequency'])
plt.xlabel('Names or Surnames')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.title('Most Common 15 Name or Surname of killed people')
plt.tight_layout()
for i, v in enumerate(name_df['frequency']):
    plt.text(i, v, str(v));
```
